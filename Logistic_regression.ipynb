modifying the data for filing the empty and na values

```python
def impute_age(col):
  age = col[1]
  pclass = col[0]
  if pd.isnull(age):
    if pclass == 1:
      return 38
    elif pclass == 2:
      return 30
    else:
      return 25
  else:
    return age
```

applying this code to df's colum named age using apply due to having two inputs as pclass and age we have to put axis = 1 also , when axis = 0 this will apply on row and code will execute but age value will be 0

```python
df['age'] = df[['pclass','age']].apply(impute_age,axis=1)
```
